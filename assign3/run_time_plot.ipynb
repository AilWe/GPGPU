```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
block_size = np.array([1, 2, 4, 8, 10, 20])
run_time = np.array([75.232, 38.816, 27.968, 37.504, 33.984, 59.776])
```

### Discussion
From the following plot, we could see that as the block size goes bigger, the execution time of kernel first goes down then goes up.

```python
plt.figure(figsize=(20,10))
plt.plot(block_size, run_time, '-o', label="Ececution Time")
plt.legend()
plt.xlabel("BLOCK_SIZE")
plt.ylabel("Time(us)")
```
